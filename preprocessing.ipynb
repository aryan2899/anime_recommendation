```python
#importing libaries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load csv 
anime_info = pd.read_csv('./data/anime.csv')
anime_synopsis = pd.read_csv('./data/anime_with_synopsis.csv')
```

```python
anime_info.head()
```

```python
anime_info.columns
```

```python
anime_info.drop(columns=['Name', 'Score', 'Genres'], inplace=True)
anime_info = pd.merge(anime_synopsis, anime_info, on='MAL_ID')
anime_info.head()
```

```python
anime_info = anime_info[['MAL_ID', 'Name', 'Score', 'Genres', 
       'Type', 'Episodes', 'Aired', 'Premiered', 'Studios', 'Source', 'Duration', 'Ranked', 'Popularity',
       'Members', 'Favorites']]
anime_info = anime_info[(anime_info['Score'] != 'Unknown')] 
```

```python
anime_info.Type.value_counts()
```

```python
types_to_drop = ['Movie', 'Special', 'OVA', 'Music']
anime_tv = anime_info[anime_info['Type'].isin(types_to_drop) == False]
anime_tv.shape
```

```python
anime_movies = anime_info[anime_info['Type'] == 'Movie']
anime_movies.shape
```
